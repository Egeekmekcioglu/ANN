```python
from numpy import exp, array, random, dot, mean, abs
import numpy
```

```python
inputx = array([[0, 0, 1], [1, 1, 1], [1, 0, 1]]) 
realresult = array([[0,1,1]]).T
```

```python
weight = array([1.0,1.0,1.0]).T
```

```python
for loop in range(1):
    SynapsValue = dot(inputx,weight)
    print("SynapsValue: \n",SynapsValue)
    predict = 1/ (1+exp(-SynapsValue))
    print("Predict: \n",predict)
    weight = weight + dot(inputx.T, ((realresult-predict)*predict*(1- predict)) )
    print("Weight: \n ",weight)
```

```python
for loop in range(1000):
    SynapsValue = dot(inputx,weight)
    predict = 1/ (1+exp(-SynapsValue))
    weight = weight + dot(inputx.T, ((realresult-predict)*predict*(1- predict)))
    print("Mean Error Func: \n",str(numpy.mean(numpy.abs(realresult-predict))))
```
